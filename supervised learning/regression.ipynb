## Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url= 'https://raw.githubusercontent.com/digipodium/Datasets/main/regression/Salary_Data.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.plot(x='YearsExperience',y='Salary',style='o')
plt.show()
```

No Preprocessing is required in this example

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df[["YearsExperience"]]     # 2D Matrix (Dataframe)
y = df["Salary"]                # 1D Array (series)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.3)
# Just for testing the result
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
acc = r2_score(y_test,y_pred)

print(f'MSE: {mse:.1f}')
print(f'MAE: {mae:.1f}')
print(f'Accuracy: {acc*100:.1f}')
```

### Visualize The Result

```python
plt.title('Linear Regression Test Result')
plt.scatter(X_test, y_test, color='grey', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2,label='Predicted')
plt.legend()
plt.text(4,11000,f'Accuracy: {acc*100:.1f}%', bbox={'edgecolor':'red'})
plt.show()
```

```python
plt.title('Linear Regression Test Result')
plt.scatter(X_test.values, y_test.values, color='grey', label='Actual')
plt.plot(X_test.values, y_pred, color='red', linewidth=2,label='Predicted')
plt.legend()
plt.text(4,11000,f'Accuracy: {acc*100:.1f}%', bbox={'edgecolor':'red'})
plt.show()
```
